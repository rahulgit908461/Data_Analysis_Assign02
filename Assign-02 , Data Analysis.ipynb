```python
#Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps
#with the highest number of installs.
```

```python
df1_copy.head(2)
```

```python
df1_copy['Installs'].value_counts()
```

```python
Installs=pd.DataFrame(df1_copy['Installs'].value_counts())
Installs.rename(columns={'Installs':'Count'},inplace=True)
```

```python
Installs
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x=Installs.index[:10],y='Count', data=Installs[:10],palette='hls')
plt.title('Top 10 App Installs')
plt.xticks(rotation=90)
plt.show()
```
